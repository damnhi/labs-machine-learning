```python
from sklearn.datasets import fetch_openml 

mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

```python
import numpy as np
print((np.array(mnist.data.loc[42]).reshape(28, 28) > 0).astype(int))
```

```python
X = mnist['data']
y = mnist['target']

y_sorted = y.sort_values()
print(y_sorted.index)

X_sorted = X.reindex(y.index)
print(X_sorted.index)

X_train_form_sorted, X_test_from_sorted = X[:56000], X[56000:]
y_train_from_sorted, y_test_sorted = y[:56000], y[56000:]
print(X_train_form_sorted.shape, y_train_from_sorted.shape)
print(X_test_from_sorted.shape, y_test_sorted.shape)
```

```python
print(f'y_train: {y_train_from_sorted.unique()}')
print(f'y_test: {y_test_sorted.unique()}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(f'y_train: {y_train.unique()}')
print(f'y_test: {y_test.unique()}')
```

```python
from sklearn.linear_model import SGDClassifier
import pickle

# Rozpoznawanie zera
y_train_0 = (y_train == '0')
y_test_0 = (y_test == '0')

sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train, y_train_0)

y_train_0_pred = sgd_clf.predict(X_train)
y_test_0_pred = sgd_clf.predict(X_test)

acc_train = sum(y_train_0_pred == y_train_0) / len(y_train_0)
acc_test = sum(y_test_0_pred == y_test_0) / len(y_test_0)

print(f'Accuracy train: {acc_train}')
print(f'Accuracy test: {acc_test}')

float_list_of_acc = [acc_train, acc_test]

with open('sgd_acc.pkl', 'wb') as f:
    pickle.dump(float_list_of_acc, f)
```

```python
from sklearn.model_selection import cross_val_score

cross_value = cross_val_score(sgd_clf, X_train, y_train_0, cv=3, scoring='accuracy')
cross_value = np.array(cross_value)

with open('sgd_cva.pkl', 'wb') as f:
    pickle.dump(cross_value, f)
```

```python
# klasyfikacja dla wszystkich cyfr
from sklearn.metrics import confusion_matrix

sgd_clf.fit(X_train, y_train)

y_train_pred = sgd_clf.predict(X_train)
y_test_pred = sgd_clf.predict(X_test)
```

```python
confusion_matrix_train = confusion_matrix(y_train, y_train_pred)
confusion_matrix_train = np.array(confusion_matrix_train)

with open('sgd_cmx.pkl', 'wb') as f:
    pickle.dump(confusion_matrix_train, f)
```
